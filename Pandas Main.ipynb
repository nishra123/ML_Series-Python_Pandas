# pandas conditions

```python
newdf.loc[(newdf['C']>0.6)]
```

returns rows that satisfy the condition that colummn C elements have value > 0.6

```python
newdf.loc[(newdf['A']<0.2) & (newdf['C']>0.6)]
```

returns rows that satisfy the condition that column A elements has value < 0.2 and column C elements have value > 0.6

**if you want to use names of rows/columns then use loc......... if you want to use indices of rows/columns then use iloc**

```python
newdf.head(2)
```

```python
newdf.iloc[0,4]
```

**inplace**

it modifies the given dataframe then and there ( jagchya jagi) when assigned inplace=True

for example if you write newdf.drop(['A', 'E'], axis=1 , inplace = True ) then it will delete columns A and E from the original dataframe newdf also

```python
newdf.drop([1,3], axis=0, inplace=True)
```

```python
newdf
```

here, rows 1 and 3 are removed!

**.drop_duplicates() is used to remove duplicates from dataframe**


here, it does't remove anything, so we will provide it a subset as:


```python
df.drop_duplicates(subset=['Name'])
```

now it removes row 3 as it had the same name as row 1

**keep = 'first' keeps first row and deletes next occuring duplicate while keep = 'second' does vice -versa ; while keep=False removes all the rows that have duplicates**

```python
df.drop_duplicates(subset=['Name'], keep='first')
```

```python
df.drop_duplicates(subset=['Name'], keep='last')
```

```python
df.drop_duplicates(subset=['Name'], keep=False)
```
